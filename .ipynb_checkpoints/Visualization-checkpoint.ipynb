# Q1: Are more unique models using alternative fuels in 2018 compared to 2008? By how much?

Checking sources of fuel and which ones are alternative sources.

```python
df_08['fuel'].value_counts()
```

```python
df_18.fuel.value_counts()
```

Looks like the alternative sources of fuel available in 2008 are CNG and ethanol, and those in 2018 ethanol and electricity.

```python
# how many unique models used alternative sources of fuel in 2008
alt_08 = df_08.query('fuel in ["CNG","ethanol"]').model.nunique()
alt_08
```

```python
# how many unique models used alternative sources of fuel in 2018
alt_18 = df_18.query('fuel in ["Electricity","Ethanol"]').nunique().model
alt_18
```

```python
plt.bar(["2008", "2018"], [alt_08, alt_18])
plt.title("Number of Unique Models Using Alternative Fuels")
plt.xlabel("Year")
plt.ylabel("Number of Unique Models");
```

```python
# total unique models each year
total_08 = df_08.model.nunique()
total_18 = df_18.model.nunique()
total_08, total_18
```

```python
prop_08 = alt_08/total_08
prop_18 = alt_18/total_18
prop_08, prop_18
```

```python
plt.bar(["2008", "2018"], [prop_08, prop_18])
plt.title("Proportion of Unique Models Using Alternative Fuels")
plt.xlabel("Year")
plt.ylabel("Proportion of Unique Models");
```

# Q2: How much have vehicle classes improved in fuel economy (increased in mpg)?

```python
veh_08 = df_08.groupby('veh_class').cmb_mpg.mean()
veh_08
```

```python
veh_18 = df_18.groupby('veh_class').cmb_mpg.mean()
veh_18
```

```python
# how much they've increased by for each vehicle class
inc = veh_18 - veh_08
inc
```

```python
# only plot the classes that exist in both years
inc.dropna(inplace=True)
plt.subplots(figsize=(8, 5))
plt.bar(inc.index, inc)
plt.title('Improvements in Fuel Economy from 2008 to 2018 by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('Increase in Average Combined MPG');
```
