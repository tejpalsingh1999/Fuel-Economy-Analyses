# Fixing Data types
    1. Fix cyl datatype
        2008: extract int from string.
        2018: convert float to int.
    
    2. Fix air_pollution_score datatype
        2008: convert string to float.
        2018: convert int to float.
    
    3. Fix city_mpg, hwy_mpg, cmb_mpg datatypes
        2008 and 2018: convert string to float.
    
    4. Fix greenhouse_gas_score datatype
        2008: convert from float to int.

## 1. Fix cyl datatype

```python
# extracting int from string of 2008 data
new_cyl = []
for data in data_2008['cyl']:
    for subdata in data.split()[0:1]:
        new_cyl.append(int(subdata[1]))
data_2008['cyl'] = new_cyl

# checking changes
data_2008.dtypes
```

```python
# converting float to int of 2018 data
data_2018 = data_2018.astype({'cyl': 'int32'})

#checking changes
data_2018.dtypes
```

## 2. Fix air_pollution_score datatypes

```python
# 2008: convert string to float
# 2008/2018 datasets contains values like '6/4' etc for hybrid fuel cars 
# splitting values into rows

# checking all the hybrids in 2008
hb_08 = data_2008[data_2008['fuel'].str.contains('/')]
hb_08
```

```python
# checking all the hybrids in 2018
hb_18 = data_2018[data_2018['fuel'].str.contains('/')]
hb_18
```

```python
# creating two copies of the 2008 hybrid dataframe
df1 = hb_08.copy() # data on first fuel type of each hybrid vehicle
df2 = hb_08.copy() # data on second fuel type of each hybrid vehicle

df1
```

```python
# selecting columns to split by '/'

split_columns = ['fuel', 'air_pollution_score', 'city_mpg', 'hwy_mpg', 'cmb_mpg', 'greenhouse_gas_score']

# applying split function to each column of each dataframe copy
for c in split_columns:
    df1[c] = df1[c].apply(lambda x : x.split('/')[0])
    df2[c] = df2[c].apply(lambda x : x.split('/')[1])
```

```python
# this dataframe holds info for the FIRST fuel type of the hybrid
# aka the values before the "/"s
df1
```

```python
# this dataframe holds info for the SECOND fuel type of the hybrid
# aka the values before the "/"s
df2
```

```python
new_rows_2008 = df1.append(df2)
new_rows_2008
```

```python
# dropping orignal hybrid rows 
data_2008.drop(hb_08.index, inplace = True)

# adding new rows to the dataframe
data_2008 = data_2008.append(new_rows_2008, ignore_index = True)

# changing the data type of 'air_pollution_score'
data_2008['air_pollution_score'] = data_2008['air_pollution_score'].astype(float)
```

```python
# checking data types
data_2008.dtypes
```

# Repeat this process for the 2018 dataset

```python
# create two copies of the 2018 hybrids dataframe, hb_18
df1 = hb_18.copy()
df2 = hb_18.copy()
```

```python
# list of columns to split
split_columns = ['fuel', 'city_mpg', 'hwy_mpg', 'cmb_mpg']

# apply split function to each column of each dataframe copy
for c in split_columns:
    df1[c] = df1[c].apply(lambda x : x.split('/')[0])
    df2[c] = df2[c].apply(lambda x : x.split('/')[1])
```

```python
# append the two dataframes
new_rows_2018 = df1.append(df2)
new_rows_2018
```

```python
# drop each hybrid row from the original 2018 dataframe
# do this by using Pandas drop function with hb_18's index
data_2018.drop(hb_18.index, inplace = True)

# append new rows to the data_2018
data_2018 = data_2018.append(new_rows_2018, ignore_index = True)
```

```python
# checking if double are gone
data_2018['fuel'].str.contains('/').all()
```

```python
# converting int to float 'air_pollution_score'
data_2018['air_pollution_score'] = data_2018['air_pollution_score'].astype(float)
```

```python
# checking changes
data_2018.dtypes
```

## 3. Fix city_mpg, hwy_mpg, cmb_mpg datatypes

```python
# convert mpg columns to floats
mpg_columns = ['city_mpg', 'hwy_mpg', 'cmb_mpg']
for c in mpg_columns:
    data_2018[c] = data_2018[c].astype(float)
    data_2008[c] = data_2008[c].astype(float)
```
