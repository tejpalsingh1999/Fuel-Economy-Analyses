# Cleaning of Data

    1. Drop extraneous columns : Drop features that aren't consistent (not present in both datasets) or aren't relevant to our questions. Use pandas' drop function.
    
    2. Rename Columns : 
        

```python
# dropping features from data_2008

data_2008 = data_2008.drop(columns = ['Stnd','Underhood ID', 'FE Calc Appr', 'Unadj Cmb MPG'])
data_2008.head(1)
```

```python
# dropping data from data_2018

data_2018 = data_2018.drop(columns = ['Stnd', 'Stnd Description', 'Underhood ID', 'Comb CO2'])
data_2018.head(1)
```

```python
# renaming 'Sale Area' column to 'Cert Region'

data_2008 = data_2008.rename(columns = {'Sales Area' : 'Cert Region'})
data_2008
```

```python
# replace spaces with underscores and lowercase labels for 2008 dataset
data_2008.rename(columns = lambda x : x.strip().lower().replace(' ','_'), inplace = True)

# confirm changes
data_2008.head(1)
```

```python
# replace spaces with underscores and lowercase labels for 2018 dataset
data_2018.rename(columns = lambda x: x.strip().lower().replace(' ','_'), inplace = True)

# confirm changes
data_2018.head(1)
```

```python
# confirm column labels for 2008 and 2018 datasets are identical
(data_2008.columns == data_2018.columns).all()
```

```python
# saving changes to new repositary 
data_2008.to_csv('data_2008_v1.csv')
data_2018.to_csv('data_2018_v1.csv')
```
